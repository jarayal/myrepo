## Part 1: The Journey - Eliza to ChatGPT

### 1966: ELIZA - The First Chatbot

How ELIZA worked:<br>
User: "I feel sad today"<br>
ELIZA: "Why do you feel sad today?"<br>


Logic:<br>
IF input contains "I feel [emotion]"<br>
THEN respond "Why do you feel [emotion]?"<br>

```python
# Simplified ELIZA logic
def eliza_response(user_input):
    """Simple funciton to implement eliza"""
    if "I feel" in user_input:
        emotion = user_input.split(" ")[-1]
        return f"Why do you feel {emotion}?"
    elif "my mother" in user_input:
        return "Tell me more about your mother"
    else:
        return "Please go on"
```

```python
eliza_response(user_input="I feel sad") 
```

```python
eliza_response(user_input="I feel happy") 
```

```python
eliza_response(user_input="I respect my mother") 
```

```python
eliza_response(user_input="I feel sad right now") 
```

```python
eliza_response(user_input="abcnd")
```
